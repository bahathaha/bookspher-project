```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_chroma import Chroma
import os
```

```python
import pandas as pd

books = pd.read_csv('books_cleaned.csv')
```

```python
books["tagged_description"]
```

```python
books["tagged_description"].to_csv("tagged_description.txt", index=False, header=False)
```

```python
from langchain_core.documents import Document

documents = []
with open("tagged_description.txt", "r", encoding="utf-8") as f:
    for line in f:
        line = line.strip()
        if line:
            documents.append(Document(page_content=line))

```

```python
documents[0]
```

```python
embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

db_books = Chroma.from_documents(
    documents,
    embedding=embeddings,
    persist_directory="./db_books"
)
```

```python
query = "A book to teach children about nature"

docs = db_books.similarity_search(query, k=10)
docs
```

```python
books[books["isbn13"] == int(docs[0].page_content.split()[0].strip().replace('"', ""))]
```

```python
def retrieve_semantic_recommendations(query: str, top_k=10)->pd.DataFrame:
    recs = db_books.similarity_search(query, k=50)

    books_list = []

    for i in range(0, len(recs)):
        books_list += [int(recs[i].page_content.strip('"').split()[0])]

    return books[books["isbn13"].isin(books_list)]
```

```python
retrieve_semantic_recommendations("Death by drowning")
```
